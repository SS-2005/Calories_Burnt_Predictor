```python
#Importing required Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
```

```python
#Uploading the actual dataset
data=pd.read_csv('/content/Dataset.csv')
```

```python
data.head()
```

```python
df=pd.DataFrame(data)
print(df)
```

```python
#Handling Missing Values
#using mean to fill in values
df['Exercise_Time_Min'] = df['Exercise_Time_Min'].fillna(df['Exercise_Time_Min'].mean())
df['Calories_Burnt']=df['Calories_Burnt'].fillna(df['Calories_Burnt'].mean())
```

```python
#Visualizeing the data set
plt.scatter(df['Exercise_Time_Min'], df['Calories_Burnt'], color='blue', label='Data Points')
plt.title('Exercise Duration vs Calories Burned')
plt.xlabel('Exercise Duration (minutes)')
plt.ylabel('Calories Burned')
plt.legend()
plt.show()
```

```python
#Spliting Data and trainig the model
X=df[['Exercise_Time_Min']] #feature-2D
Y=df['Calories_Burnt'] #target variable
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
#initializing and training the model
model=LinearRegression()
model.fit(X_train,Y_train)
```

```python
#Evaluating the model
Y_pred=model.predict(X_test)
mse=mean_squared_error(Y_test,Y_pred)
r2=r2_score(Y_test,Y_pred)
print(f"Mean Squared Error: {mse}")
print(f"R-Sqaured Score: {r2}")
```

```python
#Visualize the regression line
plt.scatter(X,Y,color='red',label='Data Points')
plt.plot(X,model.predict(X),color='blue',label='Regression line')
plt.title('Regression Line:Exercise Duration vs Calories Burnt')
plt.xlabel('Exercise Duration (minutes)')
plt.ylabel('Calories Burned')
plt.legend()
plt.show()
```

```python
#Manual Values (User Interaction)
while True:
  ui=input("Enter Exercise Duration in Minutes only \nOr type 'Exit' to quit")
  if ui.lower()=='exit':
    break
  try:
    duration=float(ui)
    pred_cal=model.predict([[duration]])[0]
    print(f"Estimate Calories Burned: {pred_cal:.2f}")
  except ValueError:
    print("Invalid Number")
```
